# My task is to demonstrate, how to plot different types of graphs.

## Introduction
Python programming is used to demonstrate how to plot different types of graphs using the popular matplotlib library. T

### This program includes examples
- **line plot**
- **bar plot**
- **scatter plot**
- **pie chart**



```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for the plots
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Line Plot
plt.figure(figsize=(10, 6))
plt.subplot(2, 2, 1)
plt.plot(x, y, label='sin(x)')
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
```

```python
# Bar Plot
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 4]

plt.subplot(2, 2, 2)
plt.bar(categories, values, color='Red')
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
```

```python
# Scatter Plot
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)

plt.subplot(2, 2, 3)
plt.scatter(x_scatter, y_scatter, color='blue')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

```python
# Pie Chart
sizes = [25, 35, 20, 20]
labels = ['Group A', 'Group B', 'Group C', 'Group D']

plt.subplot(2, 2, 4)
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['blue', 'red', 'yellow', 'green'])
plt.title('Pie Chart')

# Adjust the layout and show the plots
plt.tight_layout()
plt.show()
```
